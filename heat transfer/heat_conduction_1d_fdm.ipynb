# üî• 1D Heat Transfer in a Plate

This model describes steady-state heat conduction in a 1D plate with internal heat generation and convective boundary conditions.

---

## üìò Governing Equations

$$
\begin{cases}
\lambda \dfrac{d^2 T}{dx^2} + Q(x) = 0 & \text{for } 0 \leq x \leq E
\\[5mm]
\lambda \dfrac{dT}{dx} + H(T_d - T(x)) = 0 & \text{at } x = E
\\[5mm]
T(x) = T_d & \text{at } x = 0
\end{cases}
$$

---

## üîß Volumetric Heat Source

$$
Q(x) = Q_0 \, e^{\displaystyle \frac{x - E}{P}}
$$

**Parameters:**

- Volumetric heat generation rate: $ Q_0 \, [\mathrm{W/m^3}] $
- Characteristic heating length: $ P \, [\mathrm{m}] $

---

## üìä Given Data

$$
\begin{aligned}
&\text{Plate thickness:} && E = 0.1 \, \mathrm{m} \\
&\text{Thermal conductivity:} && \lambda = 30 \, \mathrm{W/(m \cdot K)} \\
&\text{Heat exchange coefficient:} && H = 2000 \, \mathrm{W/(m^2 \cdot K)} \\
&\text{External temperature:} && T_d = 20 \, ^\circ\mathrm{C} \\
&\text{Volumetric heat source:} && Q_0 = 5 \times 10^7 \, \mathrm{W/m^3} \\
&\text{Characteristic heating length:} && P = 0.02 \, \mathrm{m}
\end{aligned}
$$

---

## üßÆ Analytical Solution

$$
T(x) = T_d 
+ \frac{Q_0 P}{\lambda + EH} 
\left[1 + \frac{PH}{\lambda} \left(1 - e^{\displaystyle -\frac{E}{P}} \right) \right] x 
+ \frac{Q_0 P^2}{\lambda} 
e^{\displaystyle -\frac{E}{P}} \left(1 - e^{\displaystyle \frac{x}{P}} \right)
$$


# üßä Analytical Solution to 1D Steady-State Heat Conduction with Exponential Source

We solve the following boundary value problem for the temperature distribution in a 1D slab:

$$
\begin{cases}
\lambda \dfrac{d^2 T}{dx^2} + Q_0 \, e^{\displaystyle \frac{x - E}{P}} = 0 & \text{for } 0 \leq x \leq E
\\[5mm]
\lambda \dfrac{dT}{dx}(E) = H \left( T_d - T(E) \right) & \text{at } x = E
\\[5mm]
T(0) = T_d & \text{at } x = 0
\end{cases}
$$

---

## üßÆ General Solution

Start with the differential equation:

$$
\lambda \dfrac{d^2 T}{dx^2} = -Q_0 \, e^{\displaystyle \frac{x - E}{P}}
$$

Divide by $ \lambda $ and integrate once:

$$
\dfrac{dT}{dx} = -\dfrac{Q_0 P}{\lambda} \, e^{\displaystyle \frac{x - E}{P}} + C_1
$$

Integrate again:

$$
T(x) = -\dfrac{Q_0 P^2}{\lambda} \, e^{\displaystyle \frac{x - E}{P}} + C_1 x + C_2
$$

---

## üìê Apply Boundary Conditions

### At $ x = 0 $:

From the Dirichlet condition:

$$
T(0) = T_d \Rightarrow -\dfrac{Q_0 P^2}{\lambda} \, e^{\displaystyle -\frac{E}{P}} + C_2 = T_d
$$

Solve for $ C_2 $:

$$
C_2 = T_d + \dfrac{Q_0 P^2}{\lambda} \, e^{\displaystyle -\frac{E}{P}}
$$

---

### At $ x = E $:

Use the Robin-type boundary condition:

$$
\lambda \dfrac{dT}{dx}(E) = H \left( T_d - T(E) \right)
$$

Compute each side:

#### Derivative at $ x = E $:

$$
\dfrac{dT}{dx}(E) = -\dfrac{Q_0 P}{\lambda} + C_1
$$

So:

$$
\lambda \dfrac{dT}{dx}(E) = -Q_0 P + \lambda C_1
$$

#### Temperature at $ x = E $:

$$
T(E) = -\dfrac{Q_0 P^2}{\lambda} + C_1 E + C_2
$$

Substitute $ C_2 $:

$$
T(E) = -\dfrac{Q_0 P^2}{\lambda} + C_1 E + T_d + \dfrac{Q_0 P^2}{\lambda} \, e^{\displaystyle -\frac{E}{P}}
$$

Then:

$$
T_d - T(E) = -C_1 E + \dfrac{Q_0 P^2}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right)
$$

Multiply by $ H $:

$$
H \left( T_d - T(E) \right) = -H C_1 E + H \dfrac{Q_0 P^2}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right)
$$

Now equate both sides:

$$
- Q_0 P + \lambda C_1 = -H C_1 E + H \dfrac{Q_0 P^2}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right)
$$

Group terms:

$$
C_1 \left( \lambda + H E \right) = Q_0 P + H \dfrac{Q_0 P^2}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right)
$$

Solve for $ C_1 $:

$$
C_1 = \dfrac{Q_0 P}{\lambda + H E} \left( 1 + \dfrac{H P}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right) \right)
$$

---

## ‚úÖ Final Solution

Substitute $ C_1 $ and $ C_2 $ into the general solution:

$$
T(x) = T_d 
+ \dfrac{Q_0 P^2}{\lambda} \, e^{\displaystyle -\frac{E}{P}} 
\left( 1 - e^{\displaystyle \frac{x}{P}} \right) 
+ \dfrac{Q_0 P}{\lambda + H E} 
\left[ 1 + \dfrac{H P}{\lambda} \left( 1 - e^{\displaystyle -\frac{E}{P}} \right) \right] x
$$

This is the full analytical solution for the temperature profile $ T(x) $ in the domain $ [0, E] $ under steady-state conduction with an exponentially increasing source term and a Robin-type boundary condition at $ x = E $.


# üî¢ Finite Difference Method for 1D Steady-State Heat Conduction in a Plate

We discretize the following boundary value problem:

$$
\begin{cases}
\lambda \dfrac{d^2 T}{dx^2} + Q(x) = 0 & \text{for } 0 \leq x \leq E
\\[5mm]
\lambda \dfrac{dT}{dx}(E) = H(T_d - T(E)) & \text{at } x = E
\\[5mm]
T(0) = T_d & \text{at } x = 0
\end{cases}
$$

with:

$$
Q(x) = Q_0 \, e^{\displaystyle \frac{x - E}{P}}
$$

---

## üß± Step 1: Discretize the Domain

- Divide the interval $[0, E]$ into $n$ nodes: $x_1 = 0, x_2, ..., x_n = E$
- Let $ \Delta x = \dfrac{E}{n - 1} $ be the uniform grid spacing
- Denote the temperature at node $i$ by $T_i$

---

## üßÆ Step 2: Apply Central Difference to Interior Nodes

Start from:

$$
\lambda \cdot \dfrac{T_{i+1} - 2T_i + T_{i-1}}{(\Delta x)^2} + Q_0 \, e^{\frac{x_i - E}{P}} = 0
$$

Divide by $ \lambda $:

$$
\dfrac{T_{i+1} - 2T_i + T_{i-1}}{(\Delta x)^2} + \dfrac{Q_0}{\lambda} \, e^{\frac{x_i - E}{P}} = 0
$$

Rearranged:

$$
T_{i-1} - 2T_i + T_{i+1} = -(\Delta x)^2 \cdot \dfrac{Q_0}{\lambda} \, e^{\frac{x_i - E}{P}}
$$

Valid for interior nodes $i = 2, 3, ..., n - 1$

---

## üßä Step 3: Apply Dirichlet Condition at Left Boundary

At $x_1 = 0$:

$$
T_1 = T_d
$$

This value is known and substituted directly.

---

## üå°Ô∏è Step 4: Apply Robin Condition at Right Boundary

Use one-sided finite difference:

$$
\dfrac{dT}{dx} \Big|_{x_n} \approx \dfrac{T_{n+1} - T_{n-1}}{2 \Delta x}
$$

Substitute into boundary condition:

$$
\lambda \cdot \dfrac{T_{n+1} - T_{n-1}}{2 \Delta x} = H(T_d - T_n)
$$

Solve for fictitious node $T_{n+1}$:

$$
T_{n+1} = T_{n-1} + \dfrac{2 \Delta x H}{\lambda} (T_d - T_n)
$$

Substitute into FD equation at node $n$:

$$
\dfrac{T_{n+1} - 2T_n + T_{n-1}}{(\Delta x)^2} + \dfrac{Q(x_n)}{\lambda} = 0
$$

Replace $T_{n+1}$ and simplify:

$$
\dfrac{T_{n-1} + \dfrac{2 \Delta x H}{\lambda} (T_d - T_n) - 2T_n + T_{n-1}}{(\Delta x)^2} + \dfrac{Q_0}{\lambda} = 0
$$

Rearranged:

$$
-2T_{n-1} + T_n \left( 2 + \dfrac{2 \Delta x H}{\lambda} \right) = (\Delta x)^2 \cdot \dfrac{Q_0}{\lambda} + \dfrac{2 \Delta x H}{\lambda} T_d
$$

---

## üß© Step 5: Assemble the Linear System

Equations:

- For $i = 1$ (Dirichlet):

  $$
  T_1 = T_d
  $$

- For $2 \leq i \leq n - 1$ (interior):

  $$
  T_{i-1} - 2T_i + T_{i+1} = -(\Delta x)^2 \cdot \dfrac{Q_0}{\lambda} \, e^{\frac{x_i - E}{P}}
  $$

- For $i = n$ (Robin):

  $$
  2T_{n-1} - T_n \left( 2 + \dfrac{2 \Delta x H}{\lambda} \right) = -(\Delta x)^2 \cdot \dfrac{Q_0}{\lambda} - \dfrac{2 \Delta x H}{\lambda} T_d
  $$

Solve:

$$
A \cdot \mathbf{T} = \mathbf{F}
$$

Where:

- $ \mathbf{T} = [T_1, T_2, T_3, \dots, T_{n-2}, T_{n-1}, T_n]^T $
- $A$ is a tridiagonal matrix with modified first and last rows
- $\mathbf{F}$ contains source terms and boundary contributions

---

## üßÆ Matrix Form

$$
\begin{bmatrix}
1 & 0 & 0 & 0 & \cdots & 0 \\
1 & -2 & 1 & 0 & \cdots & 0 \\
0 & 1 & -2 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \ddots & \ddots & \vdots \\
0 & 0 & \cdots & 1 & -2 & 1 \\
0 & 0 & \cdots & 0 & -2 & \alpha
\end{bmatrix}
\begin{bmatrix}
T_1 \\
T_2 \\
T_3 \\
\vdots \\
T_{n-1} \\
T_n
\end{bmatrix}
=
\begin{bmatrix}
T_d \\
\gamma_2 \\
\gamma_3 \\
\vdots \\
\gamma_{n-1} \\
\beta
\end{bmatrix}
$$

---

## üìå Definitions

$$
\begin{align*}
\alpha &= 2 + \dfrac{2 \Delta x H}{\lambda}
\\[5mm]
\beta &= \dfrac{\Delta x^2}{\lambda} Q_n + \dfrac{2 \Delta x H}{\lambda} T_d
\\[5mm]
\gamma_i &= -\dfrac{\Delta x^2}{\lambda} Q_i = -\dfrac{\Delta x^2}{\lambda} Q_0 \, e^{\displaystyle \frac{x_i - E}{P}}
\end{align*}
$$

This completes the finite difference formulation and linear system setup for the 1D steady-state heat conduction problem with exponential source and mixed boundary conditions.


# üî• 1D Steady-State Heat Conduction in a Plate with Exponential Source

This notebook solves the 1D steady-state heat conduction equation using two approaches:

- ‚úÖ Analytical solution derived from the governing differential equation
- üßÆ Finite Difference Method (FDM) with mixed boundary conditions

It compares both solutions, computes error metrics, and visualizes:

- Temperature profiles
- Error distribution
- Heat generation profile

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## üîß Heat Generation Function

Defines the exponential volumetric heat source $Q(x) = Q_0 \exp\left(\frac{x - E}{P}\right)$.


```python
def heat_generation(x, Q_0, P, E):
    return Q_0 * np.exp((x - E) / P)
```

## üìê Analytical Solution

Solves the temperature profile using the closed-form expression derived from the boundary value problem.


```python
def analytical_solution(x, Q_0, Œª, H, T_d, P, E):
    term1 = T_d
    term2 = (Q_0 * P**2 / Œª) * np.exp(-E / P) * (1 - np.exp(x / P))
    term3 = (Q_0 * P / (Œª + H * E)) * (1 + (H * P / Œª) * (1 - np.exp(-E / P))) * x
    return term1 + term2 + term3
```

## üßÆ Finite Difference Method

Solves the temperature profile numerically using central difference for interior nodes and one-sided difference for the Robin boundary.


```python
def finite_difference_solution(n, dx, x, Q_0, Œª, H, T_d, P, E):
    Œ± = 2 + 2 * dx * H / Œª
    √ü = dx**2 * Q_0 / Œª + 2 * dx * H / Œª * T_d
    
    A = np.zeros((n, n))
    F = np.zeros(n)

    A[0, 0] = 1
    F[0] = T_d

    for i in range(1, n - 1):
        A[i, i - 1] = 1
        A[i, i] = -2
        A[i, i + 1] = 1
        F[i] = -dx**2 * Q_0 / Œª * np.exp((x[i] - E) / P)

    A[-1, -2] = -2
    A[-1, -1] = Œ±
    F[-1] = √ü

    return np.linalg.solve(A, F)
```

## üìâ Error Metrics

Compares the numerical and analytical solutions using pointwise error, maximum absolute error, and RMS error.


```python
def error_metrics(T_fd, T_exact):
    error = T_fd - T_exact
    max_error = np.max(np.abs(error))
    rms_error = np.sqrt(np.mean(error**2))
    return error, max_error, rms_error
```

## üìä Plot Results

Generates plots for:
- Temperature profiles
- Error distribution
- Error bars
- Heat generation profile


```python
def plot_results(x, T_exact, T_fd, error, Q_x):
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.plot(x, T_exact, label='Analytical', linewidth=2)
    plt.plot(x, T_fd, 'o--', label='Finite Difference')
    plt.xlabel('x [m]')
    plt.ylabel('Temperature [¬∞C]')
    plt.title('Temperature Distribution')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.plot(x, error, 'r-o')
    plt.xlabel('x [m]')
    plt.ylabel('Error [¬∞C]')
    plt.title('Error Distribution')
    plt.grid(True)

    plt.subplot(2, 2, 3)
    plt.errorbar(x, T_fd, yerr=np.abs(error), fmt='o', label='FD with Error')
    plt.plot(x, T_exact, label='Analytical', linewidth=2)
    plt.xlabel('x [m]')
    plt.ylabel('Temperature [¬∞C]')
    plt.title('Error Bars')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 4)
    plt.plot(x, Q_x, 'm-', linewidth=2)
    plt.xlabel('x [m]')
    plt.ylabel('Q(x) [W/m¬≥]')
    plt.title('Heat Generation Profile')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

## üöÄ Run Simulation

Defines physical parameters, computes both solutions, evaluates error, and visualizes results.


```python
E = 0.1
Œª = 30.0
H = 2000.0
T_d = 20.0
Q_0 = 5e7
P = 0.02
n_nodes = 41

dx = E / (n_nodes - 1)
x = np.linspace(0, E, n_nodes)

print("Input Parameters:")
print(f"Plate thickness E = {E} m")
print(f"Thermal conductivity Œª = {Œª} W/m¬∑K")
print(f"Heat transfer coefficient H = {H} W/m¬≤¬∑K")
print(f"External temperature T_d = {T_d} ¬∞C")
print(f"Heat generation coefficient Q_0 = {Q_0} W/m¬≥")
print(f"Characteristic length P = {P} m")
print(f"Number of nodes n = {n_nodes}")
print(f"Grid spacing dx = {dx:.5f} m\n")

T_exact = analytical_solution(x, Q_0, Œª, H, T_d, P, E)
T_fd = finite_difference_solution(n_nodes, dx, x, Q_0, Œª, H, T_d, P, E)
Q_x = heat_generation(x, Q_0, P, E)

error, max_error, rms_error = error_metrics(T_fd, T_exact)

results = pd.DataFrame({
    "x [m]": x,
    "T_exact [¬∞C]": T_exact,
    "T_FD [¬∞C]": T_fd,
    "Error [¬∞C]": error
})
print("Short Table of Results:")
print(results.head(6).to_string(index=False))
print(f"\nMax Absolute Error: {max_error:.4f} ¬∞C")
print(f"RMS Error: {rms_error:.4f} ¬∞C\n")

plot_results(x, T_exact, T_fd, error, Q_x)
```
