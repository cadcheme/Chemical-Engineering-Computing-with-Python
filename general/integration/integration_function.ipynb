### Numerical Integration of a Continuous Function Using `scipy.integrate.quad`

We compute the definite integral:

$$
S = \int_0^4 (1 - e^{-x}) \, dx
$$

This integral represents the area under the curve \( f(x) = 1 - e^{-x} \) from \( x = 0 \) to \( x = 4 \). The function starts at 0 and asymptotically approaches 1, making this a classic example of exponential saturation.

We use `scipy.integrate.quad` to evaluate the integral and visualize the function and shaded region.

**Ref**: Applied Numerical Methods with MATLABÂ® for Engineers and Scientists, Third Edition, Steven C. Chapra, Exmp 19-1 pp.492

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Define the function to integrate
def f(x):
    return 1 - np.exp(-x)

# Integration bounds
a, b = 0, 4

# Perform integration
result, error = quad(f, a, b)

# Display result
print(f"Integral of (1 - e^(-x)) from x = {a} to x = {b}:")
print(f"  S = {result:.6f}")
print(f"  Estimated error = {error:.2e}")
```

```python
# Generate x values for plotting
x_vals = np.linspace(a, b, 50)
y_vals = f(x_vals)

# Plot function and shaded area
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'r-', label='$f(x) = 1 - e^{-x}$')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= a) & (x_vals <= b),
                 color='red', alpha=0.3, label='Integration Area')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Numerical Integration of $f(x) = 1 - e^{-x}$')
plt.grid(True)
plt.legend()
plt.show()
```
