### Numerical Integration of a Parameterized Function Using `scipy.integrate.quad`

We compute the definite integral:

$$
I(a, b) = \int_0^1 (a x^2 + b) \, dx
$$

This function depends on parameters $ a $ and $ b $, which we pass using the `args` argument in `scipy.integrate.quad`.

#### Cases

- Case 1: $ a = 2, b = 1 $
- Case 2: $ a = 4, b = -2 $

We evaluate both integrals and visualize the functions over the interval $ [0, 1] $.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Function to integrate: f(x; a, b) = a*x^2 + b
def f(x, a, b):
    return a * x**2 + b
```

```python
# Integration bounds
a_int, b_int = 0, 1

# Case 1: a = 2, b = 1
I1, err1 = quad(f, a_int, b_int, args=(2, 1))

# Case 2: a = 4, b = -2
I2, err2 = quad(f, a_int, b_int, args=(4, -2))

# Display results
print("Integral Results:")
print(f"  Case 1 (a=2, b=1):     I = {I1:.6f}, error = {err1:.2e}")
print(f"  Case 2 (a=4, b=-2):    I = {I2:.6f}, error = {err2:.2e}")
```

```python
# Generate x values
x_vals = np.linspace(a_int, b_int, 300)

# Evaluate functions
y1 = f(x_vals, 2, 1)
y2 = f(x_vals, 4, -2)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1, 'b-', label='$f(x) = 2x^2 + 1$')
plt.plot(x_vals, y2, 'r-', label='$f(x) = 4x^2 - 2$')
plt.fill_between(x_vals, 0, y1, color='blue', alpha=0.2)
plt.fill_between(x_vals, 0, y2, color='red', alpha=0.2)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Parameterized Function Integration over [0, 1]')
plt.grid(True)
plt.legend()
plt.show()
```
