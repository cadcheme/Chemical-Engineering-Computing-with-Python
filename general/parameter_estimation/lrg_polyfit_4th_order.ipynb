### Fourth-Order Polynomial Regression Using NumPy

This example fits a fourth-order polynomial to pressure–temperature data using `numpy.polyfit`. The model is:

$$
P(T_C) = a_0 + a_1 T_C + a_2 T_C^2 + a_3 T_C^3 + a_4 T_C^4
$$

We estimate the coefficients $a_0$ through $a_4$ and plot the fitted curve against the original data.

#### Data

| $T_C$ (°C) | $P$ (mmHg) |
|------------|------------|
| -36.7      | 1          |
| -19.6      | 5          |
| -11.5      | 10         |
| -2.6       | 20         |
| 7.6        | 40         |
| 15.4       | 60         |
| 26.1       | 100        |
| 42.2       | 200        |
| 60.6       | 400        |
| 80.1       | 760        |

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Temperature (TC) and Pressure (P) data
TC = np.array([-36.7, -19.6, -11.5, -2.6, 7.6, 15.4, 26.1, 42.2, 60.6, 80.1])
P = np.array([1, 5, 10, 20, 40, 60, 100, 200, 400, 760])
```

```python
# Fit a 4th-order polynomial
coeffs = np.polyfit(TC, P, deg=4)

# Create polynomial function
poly4 = np.poly1d(coeffs)

# Display coefficients
print("Fourth-Order Polynomial Coefficients:")
for i, c in enumerate(coeffs[::-1]):
    print(f"  a{i} = {c:.6e}")
```

```python
# Generate smooth curve
TC_fit = np.linspace(min(TC), max(TC), 300)
P_fit = poly4(TC_fit)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(TC_fit, P_fit, color='red', label='4th Order Polynomial Fit')
plt.plot(TC, P, 'o', color='blue', label='Original Data')
plt.xlabel('Temperature $T_C$ (°C)')
plt.ylabel('Pressure $P$ (mmHg)')
plt.title('Fourth-Order Polynomial Regression: $P(T_C)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Compute predicted values and residuals
P_pred = poly4(TC)
residuals = P - P_pred

# Plot residuals
plt.figure(figsize=(8, 5))
plt.plot(TC, residuals, 'o', color='red', label='Residuals')
plt.axhline(0, color='blue', linestyle='--', label='$y=0$')
plt.xlabel('Temperature $T_C$ (°C)')
plt.ylabel('Residuals (Observed $P$ - Fitted $P$)')
plt.title('Residuals of 4th-Order Polynomial Fit')
plt.grid(True)
plt.legend()
plt.show()
```
