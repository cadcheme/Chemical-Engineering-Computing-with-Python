### System of Initial Value Problems (IVPs): Coupled Dynamic Simulation

This section demonstrates how to solve a system of two coupled ordinary differential equations (ODEs) with initial conditions using Python. These equations involve nonlinear and time-dependent terms, and are solved numerically.

#### Equations

We solve the following system:

$$
\begin{cases}
\dfrac{dx}{dt} = x + 4t^2 \\
\dfrac{dy}{dt} = \sqrt{y} - \frac{x}{5 - y}
\end{cases}
$$

with initial conditions:

$$
x(0) = 1, \quad y(0) = 4
$$

#### Purpose

To compute the time evolution of \( x(t) \) and \( y(t) \) over the interval \( t \in [0, 1] \) using `scipy.integrate.solve_ivp`. This approach is useful for modeling coupled reaction systems, population dynamics, or feedback-controlled processes.

Ref: https://polymathplus.org/

```python
# Load required libraries
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Define the system of ODEs
# dy/dt = sqrt(y) - x / (5 - y)
# dx/dt = x + 4*t^2

def system(t, vars):
    x, y = vars
    dxdt = x + 4 * t**2
    dydt = np.sqrt(y) - x / (5 - y)
    return [dxdt, dydt]
```

```python
# Initial conditions
x0 = 1.0
y0 = 4.0
t_span = (0, 1)

# Solve the system using solve_ivp
solution = solve_ivp(system, t_span, [x0, y0], dense_output=True)
```

```python
# Generate time points for plotting
t_vals = np.linspace(t_span[0], t_span[1], 200)
x_vals, y_vals = solution.sol(t_vals)

# Plot both variables
plt.plot(t_vals, x_vals, label='x(t)')
plt.plot(t_vals, y_vals, label='y(t)')
plt.xlabel('Time (t)')
plt.ylabel('State Variables')
plt.title('Coupled IVP Solution: x(t) and y(t)')
plt.grid(True)
plt.legend()
plt.show()
```
