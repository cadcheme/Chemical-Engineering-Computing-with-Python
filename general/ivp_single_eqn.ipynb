### Single Initial Value Problem (IVP): Time-Dependent Simulation

This section demonstrates how to solve a single ordinary differential equation (ODE) with an initial condition using Python. This is a foundational tool for modeling dynamic systems in chemical engineering.

#### Equation

We solve the following first-order ODE:

$$
\frac{dy}{dt} = -k y, \quad y(0) = y_0
$$

This represents exponential decay, commonly used in reaction kinetics, cooling, or radioactive decay.

#### Purpose

To compute the time evolution of \( y(t) \) using `scipy.integrate.solve_ivp`. This method is flexible and supports stiff/non-stiff systems, event detection, and adaptive time stepping.

```python
# Load required libraries
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Define the ODE function dy/dt = -k*y
# This models exponential decay with rate constant k

def decay(t, y, k):
    return -k * y
```

```python
# Define parameters
k = 0.3         # decay rate constant
y0 = [5.0]      # initial condition y(0) = 5
t_span = (0, 20)  # time interval from t=0 to t=20

# Solve the IVP using solve_ivp
solution = solve_ivp(decay, t_span, y0, args=(k,), dense_output=True)
```

```python
# Generate time points for plotting
t_vals = np.linspace(t_span[0], t_span[1], 200)
y_vals = solution.sol(t_vals)[0]

# Plot the result
plt.plot(t_vals, y_vals, label='y(t)')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Exponential Decay: dy/dt = -k*y')
plt.grid(True)
plt.legend()
plt.show()
```
