### Single Nonlinear Equation: Root Finding with Logarithmic and Polynomial Terms

This example demonstrates how to solve a nonlinear equation involving both logarithmic and quadratic terms using numerical root finding.

#### Equation

We solve:

$$
f(x) = x \cdot \log_{10}\left(\dfrac{x}{2}\right) + 4x^2 - 1620
$$

#### Interval

We search for the root in the interval:

$$
x \in [0.1,\ 100]
$$

#### Purpose

To find the value of $x$ that satisfies the equation using `scipy.optimize.root_scalar`, and to visualize $f(x)$ over the interval to identify the root location.

```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
```

```python
# Define the nonlinear function f(x)
def f(x):
    return x * np.log10(x / 2) + 4 * x**2 - 1620
```

```python
# Define x range for plotting
x_vals = np.linspace(0.1, 100, 500)
f_vals = f(x_vals)

# Plot f(x) and y=0 line
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_vals, label='$f(x)$', color='blue')
plt.axhline(0, color='red', linestyle='--', label='$y=0$')
plt.xlabel('Variable $x$')
plt.ylabel('$f(x)$')
plt.title('Root Visualization for $f(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Solve using Brent’s method
solution = root_scalar(f, bracket=[0.1, 100], method='brentq')

# Display result and test the solution
if solution.converged:
    x_root = solution.root
    residue = f(x_root)
    print(f"Root found: x = {x_root:.6f}")
    print(f"Residual at root: f(x) = {residue:.2e}")
else:
    print("Root finding did not converge.")
```
